```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('combined_demographic_data.csv')


state_stats = df.groupby('state')['demo_age_5_17'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=state_stats.values, y=state_stats.index, palette='viridis')
plt.title('Top 10 States by Age 5-17 Population')
plt.xlabel('Count')
plt.show()


plt.figure(figsize=(6,6))
sns.scatterplot(data=df, x='demo_age_5_17', y='demo_age_17_', hue='state')
plt.title('Correlation between Age Groups')
plt.show()
```
