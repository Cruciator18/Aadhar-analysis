```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plotting(dist_group):
    plt.figure(figsize=(12, 6))

    dist_group.plot(kind='bar', width=1, figsize=(15, 10))

    plt.title('Population Comparison: Age 5-17 vs 17+ by State', fontsize=14)
    plt.xlabel('State', fontsize=1)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=85)
    plt.legend(['Age 5-17', 'Age 17+'], title='Age Group')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.show()
```

```python



bio = pd.read_csv(r"C:\aadhar\api_data_aadhar_biometric\api_data_aadhar_biometric\api_data_aadhar_biometric_0_500000.csv")



bio['date'] = pd.to_datetime(bio['date'], format='%d-%m-%Y')


state_group = bio.groupby('state')[['bio_age_5_17', 'bio_age_17_']].sum()

plotting(state_group)
```
